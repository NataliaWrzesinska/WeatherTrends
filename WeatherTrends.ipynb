### 4.3 Creating a Script for Fetching Data from NOAA

```python
# Function to fetch data from the NOAA website
def fetch_noaa_data(station_id, start_year):
    base_url = "https://www.ncei.noaa.gov/access/services/data/v1"
    all_data = []
    end_date = datetime.now().strftime('%Y-%m-%d')  # Set the end date to today's date

    # Fetch data year by year
    for year in range(start_year, datetime.now().year + 1):
        params = {
            'dataset': 'daily-summaries',
            'stations': station_id,
            'startDate': f'{year}-01-01',
            'endDate': end_date if year == datetime.now().year else f'{year}-12-31',
            'format': 'csv',
            'units': 'metric'
        }
        response = requests.get(base_url, params=params)  # Send request to the server
        if response.status_code == 200:
            data = pd.read_csv(StringIO(response.text))  # Load data into a DataFrame
            all_data.append(data)  # Add data to the list
        else:
            print(f'Failed to fetch data for the year {year}')  # Inform about failure

    combined_data = pd.concat(all_data, ignore_index=True)  # Combine all data into one DataFrame
    return combined_data

# Setting parameters
station_id = "PLM00012375"
start_year = 2024

data = fetch_noaa_data(station_id, start_year)
data.to_excel("weather_data_2024.xlsx", index=False)  # Save the data to an Excel file
print("Data has been saved to weather_data_2024.xlsx")  # Confirmation of data save
```

## 6. Data Protection

The goal of data protection is to safeguard it from unauthorized access, loss, and damage. In the context of weather data analysis, data protection is essential to ensure its integrity, confidentiality, and availability.

### 6.1 Securing Data Against Unauthorized Access

In today's world, protecting data from unauthorized access is a key aspect of any project, especially when sensitive information is involved. To secure the data, appropriate encryption mechanisms and access control measures have been implemented. Below is a step-by-step process on how to encrypt data files and set access rights to ensure their security.

```python
# Generate an encryption key
key = Fernet.generate_key()
cipher_suite = Fernet(key)

# File paths
encrypted_file_path = 'weather_data_encrypted.xlsx'
key_file_path = 'encryption_key.key'

# Function to encrypt a file
def encrypt_file(input_path, output_path, cipher):
    with open(input_path, 'rb') as f:
        data = f.read()
    encrypted_data = cipher.encrypt(data)
    with open(output_path, 'wb') as f:
        f.write(encrypted_data)

# Encrypting the data file
encrypt_file(path, encrypted_file_path, cipher_suite)

# Saving the encryption key in a secure location
with open(key_file_path, 'wb') as f:
    f.write(key)

# Setting file permissions (only the owner can read and write)
os.chmod(encrypted_file_path, stat.S_IRUSR | stat.S_IWUSR)
os.chmod(key_file_path, stat.S_IRUSR | stat.S_IWUSR)

print("The file has been encrypted and saved as 'weather_data_encrypted.xlsx'.")
print("The encryption key has been saved as 'encryption_key.key'.")
print("File permissions have been set.")
```

Summary of the Security Steps:

Data Encryption and Access Control:
- Encryption and access control have been implemented to effectively protect the data from unauthorized access. The cryptography library ensures that the data is secure during both storage and transmission.

Encryption Process:
- The data file was encrypted using a key generated by the cryptography library (Fernet). The encryption key was securely saved in a separate file, and file permissions were adjusted to restrict access.

Setting File Permissions:
- Access rights were set at the operating system level, limiting read and write permissions to authorized users only. This adds another layer of security by ensuring that only the intended users can access the data.

Conclusion:

Applying these methods in the project not only protects the weather data but also meets modern data security standards, which are essential in any professional data analysis. Future actions should include regular reviews and updates of security policies to account for new threats and technologies.

### 7.1 Adding a Year and Season Column

Adding a year column and a season column allows for a more detailed analysis of weather data by considering seasonal variations. This makes it easier to identify seasonal patterns and trends in the data.

```python
# Converting the date column to datetime format
weather_data['data'] = pd.to_datetime(weather_data['data'])

# Adding a year column
weather_data['rok'] = weather_data['data'].dt.year

# Defining a function to classify seasons
def get_season(month):
    if month in [12, 1, 2]:
        return 'zima'  # Winter
    elif month in [3, 4, 5]:
        return 'wiosna'  # Spring
    elif month in [6, 7, 8]:
        return 'lato'  # Summer
    elif month in [9, 10, 11]:
        return 'jesień'  # Autumn

# Adding a season column
weather_data['sezon'] = weather_data['data'].dt.month.apply(get_season)

# Checking the first few rows of the data
print(weather_data.head())
```

### 7.2 Annual and Seasonal Analysis

To understand long-term climate changes and identify weather patterns in Poland, a detailed annual and seasonal analysis of weather data from 2014 to 2023 was conducted. This analysis covers both average temperatures and total precipitation, as well as snow cover, allowing for a comprehensive understanding of climate dynamics across different seasons.

### 7.2.1 Annual Average Temperatures

The goal of this analysis is to identify temperature trends and variability, which may reflect broader climate changes.

```python
# Calculating annual average temperature values
srednie_temperatury_roczne = weather_data.groupby('rok')['srednia_temp_c'].mean().reset_index(name='srednia_temp_roczna')
max_temperatury_roczne = weather_data.groupby('rok')['max_temp_c'].mean().reset_index(name='max_temp_roczna')
min_temperatury_roczne = weather_data.groupby('rok')['min_temp_c'].mean().reset_index(name='min_temp_roczna')

# Merging results into one table
temperatury_roczne = srednie_temperatury_roczne.merge(max_temperatury_roczne, on='rok')
temperatury_roczne = temperatury_roczne.merge(min_temperatury_roczne, on='rok')

# Displaying results
print(temperatury_roczne)

# Visualizing the results
plt.figure(figsize=(10, 6))

# Plotting annual average temperatures
plt.plot(temperatury_roczne['rok'], temperatury_roczne['srednia_temp_roczna'], marker='o', linestyle='-', label='Annual Average Temperature')
# Plotting annual maximum temperatures
plt.plot(temperatury_roczne['rok'], temperatury_roczne['max_temp_roczna'], marker='o', linestyle='-', label='Annual Maximum Temperature')
# Plotting annual minimum temperatures
plt.plot(temperatury_roczne['rok'], temperatury_roczne['min_temp_roczna'], marker='o', linestyle='-', label='Annual Minimum Temperature')

# Adding title and axis labels
plt.title('Annual Average Temperature Values (2014-2023)')
plt.xlabel('Year')
plt.ylabel('Temperature (°C)')

# Adding legend
plt.legend()

# Displaying the plot
plt.show()
```

### 7.2.2 Seasonal Average Temperatures

### 7.4 Analysis of Climate Change

The aim of this analysis is to examine long-term trends in weather data in Poland from 2014 to 2023 using linear regression and the random forest method. By analyzing these trends, potential climate changes can be detected.

### 7.4.1 Climate Change Analysis Using Linear Regression

Linear regression is a statistical technique used to model the relationship between a dependent variable and one or more independent variables. In the context of climate analysis, linear regression allows for the identification of trends in weather data over time. The slope indicates the rate of change in the dependent variable (e.g., temperature) in response to changes in the independent variable (e.g., year), while the intercept is the value of the dependent variable when the independent variable is zero.

```python
# Filtering data to exclude the year 2024
weather_data = weather_data[weather_data['rok'] < 2024]

# Calculating annual average temperature values
srednie_temperatury_roczne = weather_data.groupby('rok')['srednia_temp_c'].mean().reset_index(name='srednia_temp_roczna')

# Preparing data for regression
X = srednie_temperatury_roczne['rok'].values.reshape(-1, 1)
y = srednie_temperatury_roczne['srednia_temp_roczna'].values

# Creating a linear regression model
model = LinearRegression()
model.fit(X, y)

# Predicting values
y_pred = model.predict(X)

# Slope and intercept
slope = model.coef_[0]
intercept = model.intercept_

# Plotting linear regression results
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='orange', label='Annual Average Temperature')
plt.plot(X, y_pred, color='green', linewidth=2, label='Linear Regression')
plt.xlabel('Year')
plt.ylabel('Average Temperature (°C)')
plt.title('Trend of Annual Average Temperatures (2014-2023)')
plt.legend()
plt.grid(True)
plt.show()

# Displaying slope and intercept
print(f'Slope: {slope}')
print(f'Intercept: {intercept}')
```

```python
# Calculating annual total precipitation
suma_opadow_rocznie = weather_data.groupby('rok')['opady_mm'].sum().reset_index(name='suma_opadow_rocznie')

# Preparing data for regression
X_opady = suma_opadow_rocznie['rok'].values.reshape(-1, 1)
y_opady = suma_opadow_rocznie['suma_opadow_rocznie'].values

# Creating a linear regression model
model_opady = LinearRegression()
model_opady.fit(X_opady, y_opady)

# Predicting values
y_opady_pred = model_opady.predict(X_opady)

# Slope and intercept for precipitation
slope_opady = model_opady.coef_[0]
intercept_opady = model_opady.intercept_

# Plotting linear regression results for precipitation
plt.figure(figsize=(10, 6))
plt.scatter(X_opady, y_opady, color='blue', label='Annual Precipitation')
plt.plot(X_opady, y_opady_pred, color='green', linewidth=2, label='Linear Regression')
plt.xlabel('Year')
plt.ylabel('Total Precipitation (mm)')
plt.title('Trend of Annual Total Precipitation (2014-2023)')
plt.legend()
plt.grid(True)
plt.show()

# Displaying slope and intercept for precipitation
print(f'Slope: {slope_opady}')
print(f'Intercept: {intercept_opady}')
```

Gradual Changes:
- Although the temperature increase is visible, the rate of growth is relatively small, suggesting that changes are occurring gradually.

### 7.4.2 Climate Change Analysis Using the Random Forest Method

The Random Forest method is a machine learning technique used for classification and regression. It involves creating multiple decision trees during training and averaging the predictions of individual trees to obtain more accurate and stable results. In this analysis, the Random Forest method is applied to predict average temperatures in Poland based on historical weather data.

```python
# Converting the 'data' column to datetime
weather_data['data'] = pd.to_datetime(weather_data['data'])

# Extracting time-related features
weather_data['rok'] = weather_data['data'].dt.year
weather_data['miesiac'] = weather_data['data'].dt.month
weather_data['dzien'] = weather_data['data'].dt.day

# Selecting features and target variable
features = weather_data[['rok', 'miesiac', 'dzien']]
temperature = weather_data['srednia_temp_c']

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, temperature, test_size=0.2, random_state=42)

# Training the Random Forest model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Model evaluation
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')

# Generating dates for 2024
dates_2024 = pd.date_range(start='2024-01-01', end='2024-12-31')
features_2024 = pd.DataFrame({
    'rok': dates_2024.year,
    'miesiac': dates_2024.month,
    'dzien': dates_2024.day
})

# Predicting temperatures for 2024
predicted_2024 = model.predict(features_2024)

# Creating a DataFrame with predictions
predicted_2024_df = pd.DataFrame({
    'data': dates_2024,
    'przewidywana_srednia_temp_c': predicted_2024
})

# Aggregating historical data by month
weather_data_non_aggregated = weather_data.set_index('data')
monthly_data = weather_data_non_aggregated.resample('ME').mean(numeric_only=True)

# Aggregating predictions for 2024 by month
predicted_2024_df['miesiac'] = predicted_2024_df['data'].dt.to_period('M')
monthly_predictions_2024 = predicted_2024_df.groupby('miesiac').mean(numeric_only=True)
monthly_predictions_2024.index = monthly_predictions_2024.index.to_timestamp()

# Displaying the first few rows of predictions
print(predicted_2024_df.head(12))
```

```python
# Plotting historical and predicted average monthly temperatures
plt.figure(figsize=(14, 7))

# Historical data plot
plt.plot(monthly_data.index, monthly_data['srednia_temp_c'], label='Historical Average Temperature', color='gray', linestyle='--', marker='o')

# Predicted data plot
plt.plot(monthly_predictions_2024.index, monthly_predictions_2024['przewidywana_srednia_temp_c'], label='Predicted Average Temperature', color='orange', linestyle='-', marker='x')

# Formatting the date axis
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=3))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))

# Adding grid lines
plt.grid(visible=True, which='major', linestyle='-', linewidth='0.2', color='grey')
plt.grid(visible=False, which='minor')

# Adding titles and labels
plt.title('Historical and Predicted Average Monthly Temperature')
plt.ylabel('Average Temperature (°C)')
plt.legend()
plt.xticks(rotation=45, ha='right')

# Showing the plot
plt.tight_layout()
plt.show()

# Plotting predicted average temperatures for 2024
plt.figure(figsize=(12, 6))
plt.plot(predicted_2024_df['data'], predicted_2024_df['przewidywana_srednia_temp_c'], label='Predicted Average Temperature', color='orange')
plt.title('Predicted Average Temperature for 2024')
plt.ylabel('Average Temperature (°C)')
plt.legend()
plt.show()
```
